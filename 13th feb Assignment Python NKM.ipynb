Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.

The LookupError class is a built-in exception class in Python that serves as the base class for all errors related to accessing a collection or sequence of data. It includes a variety of specific error classes, each of which represents a different type of lookup error.

Here are two examples of errors that are subclasses of the LookupError class:

KeyError: This error is raised when we try to access a dictionary with a key that does not exist. For example:

```python
my_dict = {'a': 1, 'b': 2, 'c': 3}
value = my_dict['d']  # raises KeyError: 'd'
```

IndexError: This error is raised when we try to access a list or other sequence with an index that is out of bounds. For example:

```python
my_list = [1, 2, 3]
value = my_list[3]  # raises IndexError: list index out of range
```

Q6. List down some best practices for exception handling in python.

Some best practices for exception handling in Python:

Be specific: Use specific exception classes to catch only the types of exceptions that you expect to occur. Avoid using generic exception classes like Exception or BaseException, as these can catch more exceptions than intended and make debugging more difficult.

Keep it simple: Keep the code inside the try block as simple as possible to reduce the chance of exceptions occurring. Avoid complex operations and long chains of function calls that can make it difficult to determine where an exception occurred.

Handle exceptions appropriately: Determine the appropriate course of action for each exception that you catch. This may include logging the error, displaying a message to the user, retrying the operation, or exiting the program.

Use a finally block: Use a finally block to ensure that any resources that were opened or modified in the try block are properly cleaned up, regardless of whether an exception was raised.

Reraise exceptions selectively: If you catch an exception but cannot handle it, you can re-raise it to allow another part of the program to handle it. Use raise without an argument to re-raise the last exception that was caught.

Be cautious with bare except statements: Avoid using bare except statements, which catch all types of exceptions. Instead, catch only the specific exceptions that you expect to occur, or use a try/except block with a specific exception type and a generic except block to catch any unexpected exceptions.

Use context managers: Use context managers like with statements to ensure that resources like files, sockets, and database connections are properly closed when no longer needed.

By following these best practices, you can write more robust and maintainable code that handles exceptions appropriately and makes debugging easier.
